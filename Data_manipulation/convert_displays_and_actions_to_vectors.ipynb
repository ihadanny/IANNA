This notebook flattens displays and actions to vectors, so we can do supervised and rl learning on it

```python
import pandas as pd
import json
#import pyreact_core.query.displays as qd
#import pyreact.query.packets as qp
import operator
import json
import numpy as np
import scipy as sp

from gym_ianna.envs.ianna_env import get_state_rep_from_dicts, get_keys
```

```python

actions = pd.read_csv('experts_dataset/experts_actions.tsvv', sep = '\t', escapechar='\\')
displays = pd.read_csv('experts_dataset/experts_displays.tsv', sep = '\t', escapechar='\\')
    
```

```python
displays.head()
```

```python
all_vectors = []
def flatten_row_to_vector(row):
    data_layer = json.loads(row['data_layer'])
    granularity_layer = json.loads(row['granularity_layer'])
    res = [row['display_id']]
    res.extend(get_state_rep_from_dicts(data_layer, granularity_layer))
    all_vectors.append(res)

displays.apply(flatten_row_to_vector, axis = 1)
X = pd.DataFrame(all_vectors)
X.rename(columns={0: 'display_id'}, inplace=True)
```

```python
recs = []

def get_groupby_field_label(x):
    action = json.loads(x)
    if "groupPriority" not in action:
        return -1
    field_name = action["field"]
    field_id = get_keys().index(field_name)
    return field_id

actions.loc[:, 'group_by_field_label'] = actions.action_params.apply(get_groupby_field_label)

y = actions[actions.label != -1][['parent_display_id', 'label']]
```
